a. Nombre y apellidos de los clientes capricornianos.

b. Nombre y apellido de los clientes de nacionalidad argentina.

c. Del resultado de a) cuántos nacieron en verano.

d. Del resultado de b) quién es el cliente más joven y quién el más viejo.

```python
#Ejercicio 3_A

from datetime import datetime


def es_capricornio(fecha):
    # Formato esperado: "YYYY-MM-DD"
    mes = int(fecha[5:7])
    dia = int(fecha[8:10])
    return (mes == 12 and dia >= 22) or (mes == 1 and dia <= 19)

capricornianos = clientes.filter(lambda t: es_capricornio(t[3]))  # t[3] es la fecha de nacimiento

print(capricornianos.map(lambda t: (t[1])).take(100))
capricornianos.map(lambda t: f"{t[0]}\t{t[1]}\t{t[2]}\t{t[3]}\t{t[4]}").saveAsTextFile("/content/capricornianos")

```

# Practica 6, Ejercicio 4)

Utilizando el dataset Banco escriba un script que permita determinar si las siguientes afirmaciones son verdaderas

a) El banco tiene más clientes europeos que americanos

b) El promedio de edad de los clientes americanos es menor que el de los
europeos

c) Los americanos deben más plata que los europeos (un cliente debe plata si la suma de montos de todas sus cajas de ahorro es negativa)

d) Los clientes americanos suelen sacar, en promedio, préstamos con mayor cantidad de cuotas que los europeos

```python
#Practica 6 Ejercicio 4a

clientes = sc.textFile("/content/Clientes.txt")

def parse_nacionalidad(line):
    parts = line.split("\t")
    return parts[5]

#nacionalidades = nacionalidades_rdd.distinct().collect()
#print(nacionalidades)

def continente(nacionalidad):
    """Devuelve el continente para cada nacionalidad"""
    europa = {"ITA", "ESP"}
    america = {"BRA", "ARG", "VEN", "COL", "BOL", "ECU", "URU", "PAR", "CHI", "PER"}
    if nacionalidad in europa:
        return "Europa"
    elif nacionalidad in america:
        return "America"
    else:
        return "Otro"
nacionalidades_rdd = clientes.map(parse_nacionalidad).filter(lambda x: x is not None)

europa_count = nacionalidades_rdd.filter(lambda n: continente(n) == "Europa").count()
america_count = nacionalidades_rdd.filter(lambda n: continente(n) == "America").count()

print("Europeos: ", europa_count, "- Americanos: ", america_count)

if europa_count > america_count:
    print("✅ El banco tiene más clientes europeos que americanos")
else:
    print("❌ No tiene más europeos que americanos")
```

```python
#Practica 6 Ejercicio 4b

from datetime import date, datetime

def parse_cliente(line):
    parts = line.split("\t")
    if len(parts) >= 6:
        id_cliente = parts[0]
        nombre = parts[1]
        apellido = parts[2]
        dni = parts[3]
        fecha_nac = parts[4]
        nacionalidad = parts[5]
        return (id_cliente, nombre, apellido, dni, fecha_nac, nacionalidad)
    else:
        return None

clientes_parsed = clientes.map(parse_cliente).filter(lambda x: x is not None)
clientes = sc.textFile("/content/Clientes.txt")

def continente(nacionalidad):
    """Devuelve el continente para cada nacionalidad"""
    europa = {"ITA", "ESP"}
    america = {"BRA", "ARG", "VEN", "COL", "BOL", "ECU", "URU", "PAR", "CHI", "PER"}
    if nacionalidad in europa:
        return "Europa"
    elif nacionalidad in america:
        return "America"
    else:
        return "Otro"

#Me quedo con continente y fecha de nacimiento
clientes_continente_fecha = clientes_parsed.map(lambda x: (continente(x[5]), x[4]))

#Calcula edad a partir de fecha_nac
def calcular_edad(fecha_str):
    fecha_nac = datetime.strptime(fecha_str, "%Y-%m-%d").date()
    hoy = date.today()
    return hoy.year - fecha_nac.year - ((hoy.month, hoy.day) < (fecha_nac.month, fecha_nac.day))

# Convertir fechas a edades: (Continente, edad)
clientes_edad = clientes_continente_fecha.map(lambda x: (x[0], calcular_edad(x[1])))

# Filtrar por continente y devuelve las edades de cada continente
europa_edad_rdd = clientes_edad.filter(lambda x: x[0] == "Europa").map(lambda x: x[1])
america_edad_rdd = clientes_edad.filter(lambda x: x[0] == "America").map(lambda x: x[1])

# Promedios en base a la lista de edades
prom_europa = europa_edad_rdd.mean()
prom_america = america_edad_rdd.mean()

print(f"Edad promedio Europa: {prom_europa}")
print(f"Edad promedio América: {prom_america}")

if prom_europa > prom_america:
    print("✅ Los americanos son mas jovenes en promedio")
else:
    print("❌ Los europeos son mas jovenes en promedio")
```

# Practica 6, Ejercicio 7)

Del data set movimientos, el banco desea saber para cada uno de los meses y para cada cliente si sus movimientos tienen tendencia positiva o negativa.

a) Calcule para cada cliente y para cada mes de cada año la pendiente de la recta determinada por regresión lineal de los movimientos de dicho periodo.


```python
#Practica 6 Ejercicio 7a

from datetime import datetime

movimientos = sc.textFile("/content/Movimientos.txt")
# Formato: <ID_Caja, monto, timestamp>

def parse_movimiento(line):
    parts = line.split("\t")
    if len(parts) >= 3:
        id_caja = parts[0]
        monto = float(parts[1])
        # Updated format string to include time
        fecha_str = parts[2]
        try:
            fecha = datetime.strptime(fecha_str, "%Y-%m-%d %H:%M:%S")
        except ValueError:
            # Handle potential errors if the format is not consistent
            print(f"Skipping line due to incorrect date format: {line}")
            return None
        año = fecha.year
        mes = fecha.month
        dia = fecha.day  # lo usamos como “x” en la regresión
        return (id_caja, (año, mes, dia, monto))
    else:
        return None

movimientos_rdd = movimientos.map(parse_movimiento).filter(lambda x: x is not None)
print("movimientos:", movimientos_rdd.take(5))

cajas = sc.textFile("/content/CajasDeAhorro.txt")
# Formato: <ID_Caja, ID_Cliente, saldo>

def parse_caja(line):
    parts = line.split("\t")
    if len(parts) >= 2:
        return (parts[0], parts[1])  # (ID_Caja, ID_Cliente)
    else:
        return None

cajas_rdd = cajas.map(parse_caja).filter(lambda x: x is not None)
print("cajas", cajas_rdd.take(5))

# Asociar movimiento a cliente
movimientos_con_cliente = movimientos_rdd.join(cajas_rdd)
# (ID_Caja, ((año, mes, dia, monto), ID_Cliente))
print("movimiento + cliente", movimientos_con_cliente.take(5))


movimientos_por_cliente = movimientos_con_cliente.map(
    lambda x: ((x[1][1], x[1][0][0], x[1][0][1]), (x[1][0][2], x[1][0][3])))
# ( (ID_Cliente, año, mes), (dia, monto) )


#sumatorias para regresión lineal
def seq_op(acc, value):
    x, y = value
    return (acc[0] + x, acc[1] + y, acc[2] + x*y, acc[3] + x*x, acc[4] + 1)

def comb_op(acc1, acc2):
    return (acc1[0] + acc2[0],
            acc1[1] + acc2[1],
            acc1[2] + acc2[2],
            acc1[3] + acc2[3],
            acc1[4] + acc2[4])

# (Σx, Σy, Σxy, Σx², n)
sumatorias = movimientos_por_cliente.aggregateByKey((0,0,0,0,0), seq_op, comb_op)

def calcular_pendiente(vals):
    sum_x, sum_y, sum_xy, sum_x2, n = vals
    numerador = n * sum_xy - sum_x * sum_y
    denominador = n * sum_x2 - (sum_x ** 2)
    if denominador == 0:
        return 0  # todos los días iguales o un solo movimiento
    return numerador / denominador

pendientes = sumatorias.mapValues(calcular_pendiente)

# ========================
# 5. Clasificar tendencia
# ========================
def tendencia(p):
    if p > 0:
        return "Positiva"
    elif p < 0:
        return "Negativa"
    else:
        return "Neutra"

tendencia_por_mes = pendientes.map(lambda x: (x[0], tendencia(x[1])))
# Resultado: ((ID_Cliente, año, mes), "Positiva"/"Negativa"/"Neutra")

# ========================
# 6. Mostrar algunos resultados
# ========================
for registro in tendencia_por_mes.take(10):
    print(registro)
```
